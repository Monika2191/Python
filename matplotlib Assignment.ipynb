#### Scipy:

We have the min and max temperatures in a city In India for each months of the year.
We would like to find a function to describe this and show it graphically, the datasetgiven below.

Task:
    
1.fitting it to the periodic function

2.plot the fit

Data

Max = 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25

Min = 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize
```

```python
Max_temp =np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
Min_temp =np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])
months=np.arange(12)
#defining periodic function
def yearly_temps(times, avg, ampl, time_offset):
    return (avg + ampl * np.cos((times + time_offset) * 1.8 * np.pi / times.max()))
res_max, cov_max = optimize.curve_fit(yearly_temps, months,Max_temp, [40, 20,0])
res_min, cov_min = optimize.curve_fit(yearly_temps, months,Min_temp, [-40, 20,0])
#plot
days = np.linspace(0, 12, num=365)
plt.figure()
plt.plot(months,Max_temp, 'ro')
plt.plot(days, yearly_temps(days, *res_max), 'r-')
plt.plot(months,Min_temp, 'bo')
plt.plot(days, yearly_temps(days, *res_min), 'b-')
plt.xlabel('Month')
plt.ylabel('Temperature (Â°C)')
plt.show()
```

Charts to plot:

1. Create a pie chart presenting the male/female proportion

```python
url="https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv"
titanic = pd.read_csv(url,skip_blank_lines=True,skipfooter=1,engine='python')
titanic.head()
```

```python
titanic.sex.unique()
```

```python
titanic.sex.value_counts()
```

```python
fig, axis = plt.subplots()
plt.pie(titanic.sex.value_counts(), labels=["Male","Female"], colors=["b","r"],autopct='%1.1f%%')
plt.title("Male/Female Proportion")
plt.show()
```

2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

```python
fig,axis=plt.subplots()
colors = {'male':'b', 'female':'r'}

axis.scatter(titanic.fare, titanic.age, alpha=0.5, c=titanic['sex'].apply(lambda x: colors[x]))
axis.set_xlabel('Fare')
axis.set_ylabel('Age')
axis.set_title('Fare paid and the Age')
```
